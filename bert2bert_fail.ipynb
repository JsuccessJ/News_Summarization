```python
encoder_max_length = 512
decoder_max_length = 128

def process_data_to_model_inputs(batch):

    # Tokenize the inputs and labels
    inputs = tokenizer(list(batch["original"]), padding="max_length", truncation=True, max_length=encoder_max_length)
    outputs = tokenizer(list(batch["extractive"]), padding="max_length", truncation=True, max_length=decoder_max_length)

    # Assign tokenized values to the batch dictionary
    batch["input_ids"] = inputs.input_ids
    batch["attention_mask"] = inputs.attention_mask
    batch["decoder_input_ids"] = outputs.input_ids
    batch["decoder_attention_mask"] = outputs.attention_mask

    # Prepare labels, ensuring that the PAD token is ignored
    batch["labels"] = [[-100 if token == tokenizer.pad_token_id else token for token in labels] for labels in outputs.input_ids]

    return batch
```

```python
from datasets import Dataset
# 데이터프레임을 데이터셋으로 변환
custom_dataset = Dataset.from_pandas(df)

# 매핑 함수를 사용하여 데이터셋 전처리
processed_dataset = custom_dataset.map(
    process_data_to_model_inputs,
    batched=True,
    batch_size=4,  # 적절한 배치 크기로 설정
    remove_columns=["original","extractive","__index_level_0__"]
)
```

```python
from transformers import Seq2SeqTrainer
from transformers import Seq2SeqTrainingArguments
```

```python
training_args = Seq2SeqTrainingArguments(
    predict_with_generate=True,           # 모델 생성을 사용하여 예측 수행
    evaluation_strategy="steps",          # 평가 전략: 훈련 스텝 기준
    per_device_train_batch_size=8,       # 각 기기당 훈련 배치 크기
    per_device_eval_batch_size=8,        # 각 기기당 평가 배치 크기
    output_dir="./",                      # 모델과 로그를 저장할 디렉토리
    logging_steps=2,                      # 몇 스텝마다 로깅할지 설정
    save_steps=10,                        # 몇 스텝마다 모델 저장할지 설정
    eval_steps=4,                         # 몇 스텝마다 평가할지 설정
    # logging_steps=1000,
    # save_steps=500,
    # eval_steps=7500,
    # warmup_steps=2000,
    # save_total_limit=3,
)
```

```python
import datasets
rouge = datasets.load_metric("rouge")
```

```python
def compute_metrics(pred):
    labels_ids = pred.label_ids
    pred_ids = pred.predictions

    pred_str = tokenizer.batch_decode(pred_ids, skip_special_tokens=True)
    labels_ids[labels_ids == -100] = tokenizer.pad_token_id
    label_str = tokenizer.batch_decode(labels_ids, skip_special_tokens=True)

    rouge_output = rouge.compute(predictions=pred_str, references=label_str, rouge_types=["rouge2"])["rouge2"].mid

    return {
        "rouge2_precision": round(rouge_output.precision, 4),
        "rouge2_recall": round(rouge_output.recall, 4),
        "rouge2_fmeasure": round(rouge_output.fmeasure, 4),
    }
```

```python
# instantiate trainer
trainer = Seq2SeqTrainer(
    model=bert2bert,
    args=training_args,
    compute_metrics=compute_metrics,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    
)
```

```python
trainer.train()
```
